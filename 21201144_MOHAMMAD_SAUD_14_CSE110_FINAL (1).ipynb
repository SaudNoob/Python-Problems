Question-1:\
Assume, you have been given two lists. Your task is to multiply the first 
element of the first list, to the last element of the second list and store 
them in another list. Then, again multiply the second element of the first 
list, to the second last element of the second list, and so on.

If any of your lists is out of element, and another list has some elements 
remaining, then just append the remaining elements to your final list and 
print the list.\
================================================

Given lists 1:\
list_one = [1, 4, 7, 5]\
list_two = [6, 1, 3, 9]

Sample Output 1:\
[9, 12, 7, 30]

Explanation 01:\
Here, from list_one’s 1st element, 1 has been multiplied 
with the last element of list_two 9. So, the product 1X9 =9 has been added 
to the answer.  Then, from list_one’s 2nd element, 4 has been multiplied 
with the 2nd the last element of list_two 3. So, the product 3X4 =12 has 
been added to the answer. And so on.\

================================================
Given lists 2:\
list_one = [1, 4, 7, 5 , 8, 2 , 5 ]\
list_two =  [6, 1, 3, 9]

Sample Output 2:\
[9 , 12, 7, 30, 8, 2, 5 ]

Explanation 02:\
Here, list_one has additional elements  8, 2 , 5  at the end. So, after 
adding the multiplied values, the additional elements have been appended to 
the answer list.

================================================
Given lists 3:\
list_one = [6, 1, 3, 9]\
list_two =  [1, 4, 7, 5 , 8, 2 , 5 ]

Sample Output 3:\
[30, 2, 24, 45, 1, 4, 7]

Explanation 03:\
Here, list_two has additional elements  1, 4, 7,  at the beginning. So, 
after adding the multiplied values, the additional elements have been 
appended to the answer list.

```python
list1 = [int(x) for x in input()[1:-1].split(',')]
list2 = [int(y) for y in input()[1:-1].split(',')]
lst = []
sm = 0
if len(list2) > len(list1):
    for i in range(len(list1)):
            sm = list1[i]*list2[len(list2)-1-i]
            lst.append(sm)
    leftlist2 = list2[0:len(list1)-1]
    for j in leftlist2:
        lst.append(j)
elif len(list2) == len(list1):
    for i in range(len(list1)):
            sm = list1[i]*list2[len(list2)-1-i]
            lst.append(sm)    
elif len(list2) < len(list1):
    for i in range(len(list2)):
            sm = list1[i]*list2[len(list2)-1-i]
            lst.append(sm)
    leftlist1 = list1[len(list2):]
    for j in leftlist1:
        lst.append(j)
print('Output list:')
print(lst)
```

Question-2:\
Assume, you have been given a string that contains a few products name, 
prices, and, costing respectively, and an integer value as a budget for 
shopping.

Now, write a python program that will calculate the total costing of 
products from the given string and check whether it is possible to buy 
those products while being within the budget.
If the total costing is greater than the budget, then print “Shopping not 
possible”.\
Otherwise, print  “Yay!! Shopping possible”.
If the quantity of any of the products is less than 1, then print “Quantity 
must be greater than 0”.

======================================================\
Given 1:\
data = "Apples,5,2,Oranges,10,1,Eggs,3,5"\
budget= 70

Sample Output 1:\
Costing in dictionary: {'Apples': 10, 'Oranges': 10, 'Eggs': 15}\
Total costing: 35\
Total Budget: 70\
Yay!! Shopping possible

Explanation 1:\
The 1st product is Apples, and each apple costs 5 takas and 2 is the buying 
quantity of apples. So, the cost of apples is 5*2=10 taka.
The 2nd product is Oranges, and each orange costs 10 takas, and 1 is the 
buying quantity of oranges. So, the cost of Orange is 10*1=10 taka.
The 3rd product is Eggs, and each egg costs 3 takas and 5 is the buying 
quantity of eggs. So, the cost of eggs is 3*5=15 taka.
So, total costing= 10+10+15=35 taka which is less than the budget of 70. 
So, You can buy these items.\

The 3rd product is Eggs, and each egg costs 3 taka and 5 is the buying

```python
a_string = input('Enter Data: ')[1:-1]
print()
budget= 70
dic1 = {}
lst = a_string.split(',')
x = []
y = []
flag = False
for num in lst:
    if num.isdigit():
        x.append(int(num))
    else:
        y.append(num)
# print(x,y)    
lstx = []
for i in range(0,len(x),2):
    sm = x[i]*x[i+1]
    lstx.append(sm)
#print(lstx)
if 0 not in lstx:
    for j in range(len(y)):
        dic1[y[j]] = lstx[j]
    print('Costing in dictionary: ',dic1)
    total = 0
    for h in lstx:
        total += h
    print('Total costing:',total)
    print('Total Budget:',budget)
    if total <= budget:
        print('Yay!! Shopping possible')
    else:
        print('Shopping not possible')
else: 
    print('Quantity must be greater than 0')
```

```python
def take_order1():
    price_list={'Juice':30, 'Fried Rice':150, 'Burger':50, 'Pizza':110, 'Fried Chicken':90, 'Noodles':70}
    lst = []
    lst2 = []
    sm = 0
    print('Please enter the name of the items:')
    while True:
        order = input()
        if order == "X":
            break 
        else:
            lst.append(order)
    for i in lst:
        if i in price_list:
            lst2.append(price_list[i])
    for j in lst2:
        sm += j
    return sm
FrstC = take_order1()
print('Total cost:',FrstC)
print()
def take_order2():
    price_list={'Juice':30, 'Fried Rice':150, 'Burger':50, 'Pizza':110, 'Fried Chicken':90, 'Noodles':70}
    lst = []
    lst2 = []
    sm = 0
    print('Please enter the name of the items:')
    while True:
        order = input()
        if order == "X":
            break 
        else:
            lst.append(order)
    for i in lst:
        if i in price_list:
            lst2.append(price_list[i])
    for j in lst2:
        sm += j
    return sm
secdC = take_order2()
print('Total cost:',secdC)
print()
def take_order3():
    price_list={'Juice':30, 'Fried Rice':150, 'Burger':50, 'Pizza':110, 'Fried Chicken':90, 'Noodles':70}
    lst = []
    lst2 = []
    sm = 0
    print('Please enter the name of the items:')
    while True:
        order = input()
        if order == "X":
            break 
        else:
            lst.append(order)
    for i in lst:
        if i in price_list:
            lst2.append(price_list[i])
    for j in lst2:
        sm += j
    return sm
thrdC = take_order3()
print('Total cost:',thrdC)
print()
Total = FrstC + secdC + thrdC
print('Total income of the day:',Total)
if Total > 500:
    print('Great Work')
else:
    print('Do Best Next Day.')
```
